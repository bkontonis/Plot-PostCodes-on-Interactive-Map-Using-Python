```python
import pgeocode
import plotly.express as px
import pandas as pd
pd.set_option("display.max_rows", None)
```

```python
nomi = pgeocode.Nominatim('de')
post_df = pd.read_csv("./postcodes.csv", sep=";")
```

```python
post_df = post_df.iloc[:,1].to_frame()
```

```python
post_df.shape
```

```python
# Test example
nomi.query_postal_code("37085")
```

```python
# Test example - Extract Latitude
nomi.query_postal_code("37085")[9]
```

```python
post_df["latitude"] = post_df["Customer PostCode"].apply(lambda x: nomi.query_postal_code(x)[9])
post_df["longitude"] = post_df["Customer PostCode"].apply(lambda x: nomi.query_postal_code(x)[10])
```

```python
post_df.head()
```

```python
fig = px.scatter_mapbox(post_df, lat="latitude", lon="longitude", #hover_name="City", hover_data=["State", "Population"],
                        #color_discrete_sequence=["fuchsia"], 
                        zoom=3, height=300)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
